```python
N, K = input().split()

N = int(N)
K = int(K)

count = 0
i = 1 # 1부터 나누기 시작
list = []


while i <= N: # i가 N보다 같거나 작을 때까지만 나누기
  if N % i == 0: # 0이면 약수라고 판단
    count += 1 # count 하나 올려주고
    list.append(i) # list에 넣어준다.
  i += 1

if (count < K): # count가 K보다 적게 되면 0 출력
  print(0)
else: # 아니면 K번째 작은 수, 즉 뒤에서 두 번째
  print(list[K-1])
```
